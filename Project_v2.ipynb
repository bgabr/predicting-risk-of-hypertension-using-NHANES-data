```python
def combinedata(_2017_2020,_2015_2016,_2013_2014,_2011_2012,_2009_2010,_2007_2008):
    _17_20_list = _2017_2020.columns.tolist()
    _15_16_list = _2015_2016.columns.tolist()
    _13_14_list = _2013_2014.columns.tolist()
    _11_12_list = _2011_2012.columns.tolist()
    _09_10_list = _2009_2010.columns.tolist()
    _07_08_list = _2007_2008.columns.tolist()
    
    intersect_col = set(_17_20_list).intersection(_15_16_list).intersection(_13_14_list).intersection(_11_12_list).intersection(_09_10_list).intersection(_07_08_list)
    
    _17_20_filtered = pd.DataFrame(_2017_2020, columns=intersect_col)
    _15_16_filtered = pd.DataFrame(_2015_2016, columns=intersect_col)
    _13_14_filtered = pd.DataFrame(_2013_2014, columns=intersect_col)
    _11_12_filtered = pd.DataFrame(_2011_2012, columns=intersect_col)
    _09_10_filtered = pd.DataFrame(_2009_2010, columns=intersect_col)
    _07_08_filtered = pd.DataFrame(_2007_2008, columns=intersect_col)
    print(intersect_col)
    
    _final = _17_20_filtered.append(_15_16_filtered)
    _final = _final.append(_13_14_filtered)
    _final = _final.append(_11_12_filtered)
    _final = _final.append(_09_10_filtered)
    _final = _final.append(_07_08_filtered)
    
    return _final
```

### Feature selection on categorical variables

#### Correlation between categorical variables to the response varaible

Comment: I will be using chi square test to test the correlation between 2 categorical variables (1 predictor and 1 response variable). The chisquare function involves:

- Assumption H0: the features are not related to each other
- Assumption H1: the features are related to each other
- The function returns the p-value, if p-value is high, we don't reject H0 and if p-value is below 0.05 then we accept H1.

Reference: https://thinkingneuron.com/how-to-measure-the-correlation-between-two-categorical-variables-in-python/

```python
from sklearn.impute import SimpleImputer
from scipy.stats import chi2_contingency
import numpy as np

def categorical_correlation(feature):
    print(feature)
    test_feat = final_df_2[[feature]]
    imp_test = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    test_feat = imp_test.fit_transform(test_feat)
    test_feat = pd.DataFrame(test_feat, columns=[feature])
    CrosstabResult=pd.crosstab(index=test_feat[feature],columns=final_df_2['BPQ020'])
    print("Cross table result:")
    print(CrosstabResult)
    ChiSqResult = chi2_contingency(CrosstabResult)
    
    print("\nchisquare p-value:")
    print(ChiSqResult[1])
    print("\n")
```

```python
for i in range(len(categorical_features)):
    categorical_correlation(categorical_features[i])
```

#### Correlation between numeric variables and the response variable

Comment: to test the correlation between the numeric features and the response variable (binary), i will be using the pointbiserial correlation method.

Reference: https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.pointbiserialr.html#:~:text=The%20point%20biserial%20correlation%20is,1%20imply%20a%20determinative%20relationship.

```python
# filling in all the missing values in the median using simple imputer.
imp_numeric = SimpleImputer(missing_values=np.nan, strategy='median')
numeric_feats_transformed = imp_numeric.fit_transform(numeric_features)
numeric_feats_transformed = pd.DataFrame(numeric_feats_transformed, columns=numeric_features.columns)
numeric_feats_transformed.isnull().sum()

numeric_features = numeric_feats_transformed.columns

def numeric_correlation(feature):
    print("feature: ", feature)
    y = final_df_2['BPQ020']
    result = stats.pointbiserialr(numeric_feats_transformed[feature], y)
    print("Correlation:")
    print(result[0])
    print("p-value:")
    print(result[1])
    print("\n")
    
for i in range(len(numeric_features)):
    numeric_correlation(numeric_features[i])
```

```python
numeric_features = hyp_nhanes_numeric_features_transformed.columns

def numeric_correlation(feature):
    print("feature: ", feature)
    y = hyp_nhanes_response_feature['hypertension']
    result = stats.pointbiserialr(hyp_nhanes_numeric_features_transformed[feature], y)
    print("Correlation:")
    print(result[0])
    print("p-value:")
    print(result[1])
    print("\n")
    
for i in range(len(numeric_features)):
    numeric_correlation(numeric_features[i])
```

#### Relationship between categorical variable and response variable

```python
categorical_features = hyp_nhanes_categorical_features_transformed.columns

def categorical_correlation(feature):
    CrosstabResult=pd.crosstab(index=hyp_nhanes_categorical_features_transformed[feature],
                               columns=hyp_nhanes_response_feature['hypertension'])
    print("Cross table result:")
    print(CrosstabResult)
    ChiSqResult = chi2_contingency(CrosstabResult)
    
    print("\nchisquare p-value:")
    print(ChiSqResult[1])
    print("\n")
    
    
for i in range(len(categorical_features)):
    categorical_correlation(categorical_features[i])
```

Comment: from the table above, it appears that race, gender, mod_act, are related to the response variable since their p-values are very small.
